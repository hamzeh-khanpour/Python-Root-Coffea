```python
import matplotlib.pyplot as plt
import numpy as np

# Example data for each group
labels = ['NNLO (Fit D)', 'NNLO (Fit D + EIC)', 'NNLO (Fit D + EIC + jet/dijet)', 'NNPDF3.1', 'MMHT14', 'ATLAS-epWZ16']
x_values = [1.069, 1.164, 1.138, 1.05, 1.25, 1.19]  # Central values for each dataset
x_err = [0.053, 0.051, 0.048, 0.12, 0.23, 0.07]    # Error values

# Define markers and colors for each dataset
markers = ['s', 'P', '^', '*', 'o', 'o']
colors = ['black', 'teal', 'red', 'orange', 'blue', 'purple']
fill_styles = ['full', 'full', 'full', 'full', 'full', 'none']  # One marker is not filled

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))

# Plot each dataset with error bars
for i, label in enumerate(labels):
    ax.errorbar(x_values[i], i, xerr=x_err[i], fmt=markers[i], color=colors[i],
                markersize=10, label=label, capsize=5, capthick=2, elinewidth=2,
                markerfacecoloralt='white' if fill_styles[i] == 'none' else colors[i],
                fillstyle=fill_styles[i])

# Customize the plot
ax.set_yticks(np.arange(len(labels)))    # Set y-ticks as labels
ax.set_yticklabels(labels, fontsize=14)  # Add dataset names on y-axis
ax.set_xlabel(r'$r_s$', fontsize=16)     # Set x-axis label with LaTeX formatting
ax.set_title(r'$Q^2 = 1.9 \, \mathrm{GeV}^2, x = 0.023$', fontsize=18)  # Title with LaTeX

# Invert y-axis to match the example plot
ax.invert_yaxis()

# Add legend on the right side
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14)

# Tight layout and save as PDF
plt.tight_layout()
plt.savefig('rs_comparison_plot.pdf', format='pdf')

# Show the plot
plt.show()
```
